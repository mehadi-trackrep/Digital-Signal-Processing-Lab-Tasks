```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def sinWave(a,T,t):
    f = 1/T
    y=a*np.sin(2*np.pi*f*t)
    return y
```

```python
t=np.arange(0.0,100,0.01)
y=sinWave(5,10,t)
plt.plot(t,y, 'm')
plt.show()
```

### Task#04: Plot the basic signals like unit impulse, unit samples, unit step, exponential, unit ramp (discussed in classes) etc.

```python
fig = plt.figure(figsize= [20,10])
###Unit Impulse
t=np.arange(-20,20,1)
y=[]
mid=0
for x in t:
    if x==mid:
        y.append(1)
    else:
        y.append(0)
plt.subplot(2,2,1)      
plt.title("Unit Impulse")
plt.stem(t,y)

###unit Step
t1=np.arange(-20,20,1)
y1=[]
for x in t1:
    if x>=0:
        y1.append(1)     
    else:
        y1.append(0)
plt.subplot(2,2,2)       
plt.title("Unit Step")
plt.stem(t1,y1)

###unit Exponential
t2=np.arange(-20,20,0.1)
y2=[]
for x in t2:
    y2.append(np.exp(x))
plt.subplot(2,2,3)       
plt.title("Exponential")
plt.plot(t2,y2)

###unit ramp
t3=np.arange(-20,20,0.1)
y3=[]
for x in t3:
    if x>=0:
        y3.append(x)
    else:
        y3.append(0)
plt.subplot(2,2,4)       
plt.title("Ramp")
plt.stem(t3,y3, 'c')

plt.show()
```

### Task#05: Sampling of a Signal... take a signal then sample using a sampling rate.

```python
def SubPloting(nrows,ncols,index,t,y,color,title,xlabel,ylabel,markersize,show):
    plt.subplot(nrows,ncols,index)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    l=plt.plot(t,y,color)
    if markersize > 0:
        plt.setp(l,markersize=markersize)
    if show==1:
        plt.show()
```

```python
fig = plt.figure(figsize= [20,10])
T=10
f=1/T
fs=2*f

t=np.arange(0.0, 50, 0.01)
y=sinWave(5,10,t)
SubPloting(2,2,1,t,y,'b',"Main Signal",'','',0,0)

# fs = fs-0.05
fs = fs+1
tdif = 1/fs
t=np.arange(0.0, 50, tdif)
y=sinWave(5,T,t)
markersize=10
SubPloting(2,2,2,t,y,'ro',"After Sampling",'','',markersize,1)
```

### Task#06: Sampling Theorem Experiment.

```python
fig = plt.figure(figsize= [20,10])
T=10
f=1/T
fs=2*f

t=np.arange(0.0, 50, 0.01)
y=sinWave(5,10,t)
SubPloting(2,2,1,t,y,'b',"Main Signal",'Time =>','Amplitue =>',0,0)

# fs = fs-0.05
fs = fs+0.1
tdif = 1/fs
t=np.arange(0.0, 50, tdif)
y=sinWave(5,T,t)
markersize=10
SubPloting(2,2,2,t,y,'ro',"After Sampling",'Time =>','Amplitue =>',markersize,1)
```

### Task#07: Implement some basic systems like Adder, Multiplier, Delay, Advanced, Fold etc.

```python
def Adder(x1,x2):
    l = len(x1)
    y = []
    for i in range(l):
        y.append(x1[i] + x2[i])
    return y

        
def Multiplier(mul, x):
    l = len(x)
    y = []
    for i in range(l):
        y.append( (x[i]) * mul )
    return y

def Delay(d, x):
    l = len(x)
    y = []
    for i in range(l):
        y.append(x[i])
    for i in range(d):
        y.append(0)
    return y

def Advance(a, x):
    l = len(x)
    y = []
    for i in range(a):
        y.append(0)
    for i in range(l):
        y.append(x[i])
    return y

def Fold(x):
    return x[::-1]

x1 = [1,2,3,4]
x2 = [4,3,2,1]
x  = [1,2,3,4]
d = 2
mul = 5
a = 2

print("Adder: ",Adder(x1,x2))
print("Multiplier: ",Multiplier(mul, x))
print("Delay: ",Delay(d, x))
print("Advance: ",Advance(a, x))
print("Fold: ",Fold(x))
```

### Task#08: Implement Moving average filter, Median Filter, Accumulator etc.

```python
import os
import sys
import seaborn
from scipy.signal import medfilt
```

```python
def movingaverage(values,window):
    weigths = np.repeat(1.0, window)/window
    m = np.convolve(values, weigths, 'valid')
    #print ("m: ",m)
    return m

signal = [1,6,3,13,5,6,7,10,9,4,6,5,7]

plt.subplot(2,2,1)       
plt.title("Moving Average")
plt.plot(movingaverage(signal,3), 'r-')
plt.show()

def median_filter(data):
    return medfilt(data, kernel_size=3)

plt.subplot(2,2,1)       
plt.title("Median Filter")
plt.plot(median_filter(signal), 'r-')
plt.show()
```

### Task#10: program to find the DFT/IDFT

```python
import numpy.fft as DFT ### Getting DFT using fft
N = 100
f=1/10
t = np.linspace(0,20,N)
x = np.sin(2*np.pi*f*t)

y = DFT.fftn(x, norm='ortho')
window = np.zeros(N)
window[:20] = 1
yr = DFT.ifftn(y*window, norm='ortho')

plt.plot(t, x, '-bx')
plt.plot(t, yr, 'ro')
plt.show()
```

### Tasks#11: DFT/IDFT -> Composite Sinusoidal Signal to see the frequencies

```python
import numpy.fft as DFT ### Getting DFT using fft
N = 100
f=1/10
t = np.linspace(0,20,N)
x = np.sin(2*np.pi*f*t) + np.exp(-t/3)*np.cos(2*np.pi*f*t) ##Composite signal

y = DFT.fftn(x, norm='ortho')
window = np.zeros(N)
window[:20] = 1
yr = DFT.ifftn(y*window, norm='ortho')

plt.plot(t, x, '-bx')
plt.plot(t, yr, 'ro')
plt.show()
```

### Task#13:  Linear Convolution

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig

L = 32  # length of signal x[k]
N = 16  # length of signal h[k]
M = 16  # periodicity of periodic convolution


def cconv(x, h, M):
    L = len(x)
    N = len(h)
    # periodic continuation of x
    xc = np.copy(x)
    if M<L:
        xc[0:-M+L] += xc[M:L]
        xc = xc[0:M]
    # zero-padding of h
    hp = h
    if N<M:
        hp = np.append(hp, np.zeros(M-N))
    # circular convolution
    y = [np.dot(np.roll(xc[::-1], k+1), hp) for k in np.arange(M)]

    return np.asarray(y)

# generate signals
x = np.ones(L)
h = sig.triang(N)

# linear convolution
y1 = np.convolve(x, h, 'full')

# plot results
def plot_signal(x):
    plt.figure(figsize = (10, 3))
    plt.stem(x)
    plt.xlabel(r'$k$')
    plt.ylabel(r'$y[k]$')
    plt.axis([0, N+L, 0, 1.1*x.max()])

plot_signal(x)
plt.title('Signal $x[k]$')

plot_signal(y1)
plt.title('Linear convolution')
```

### Task#14: Circular Convolution Program

```python
# periodic convolution
y2 = cconv(x, h, M)

plot_signal(y2)
plt.title('Periodic convolution')
```

### Task#15: Linear Convolution by FFT, IFFT [DFT, IDFT]

```python

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig

L = 16  # length of signal x[k]
N = 16  # length of signal h[k]
M = N+L-1

# generate signals
x = np.ones(L)
h = sig.triang(N)

# fast convolution
y2 = np.fft.ifft(np.fft.fft(x, M)*np.fft.fft(h, M))

plt.figure(figsize=(10, 3))
plt.stem(y2)
plt.xlabel(r'k')
plt.ylabel(r'|e[k]|');
```

### Task#16: Circular Convolution by FFT, IFFT [DFT, IDFT]

```python

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig

L = 16  # length of signal x[k]
N = 16  # length of signal h[k]
M = N+L-1

# generate signals
x = np.ones(L)
h = sig.triang(N)

def circular_conv():
    return np.fft.ifft(np.fft.fft(x, M)*np.fft.fft(h, M))

y = circular_conv()

plt.figure(figsize=(10, 3))
plt.title("-- Circular Convolution using FFT, IFFT --")
plt.stem(y)
plt.xlabel(r'k')
plt.ylabel(r'|e[k]|');
```

### Task17# FIR Filter Low Pass

```python
from pylab import *
import scipy.signal as signal

#Plot frequency and phase response
def mfreqz(b,a=1):
    w,h = signal.freqz(b,a)
    h_dB = 20 * log10 (abs(h))
    subplot(211)
    plot(w/max(w),h_dB)
    ylim(-150, 5)
    ylabel('Magnitude (db)')
    xlabel(r'Normalized Frequency (x$\pi$rad/sample)')
    title(r'Frequency response')
    subplot(212)
    h_Phase = unwrap(arctan2(imag(h),real(h)))
    plot(w/max(w),h_Phase)
    ylabel('Phase (radians)')
    xlabel(r'Normalized Frequency (x$\pi$rad/sample)')
    title(r'Phase response')
    subplots_adjust(hspace=0.5)

#Plot step and impulse response
def impz(b,a=1):
    l = len(b)
    impulse = repeat(0.,l); impulse[0] =1.
    x = arange(0,l)
    response = signal.lfilter(b,a,impulse)
    subplot(211)
    stem(x, response)
    ylabel('Amplitude')
    xlabel(r'n (samples)')
    title(r'Impulse response')
    subplot(212)
    step = cumsum(response)
    stem(x, step)
    ylabel('Amplitude')
    xlabel(r'n (samples)')
    title(r'Step response')
    subplots_adjust(hspace=0.5)
n = 61
a = signal.firwin(n, cutoff = 0.3, window = "hamming")
#Frequency and phase response
mfreqz(a)
```

### Task20# IIR Filter Low Pass

```python

from numpy import *
from scipy import signal
from scipy.signal import filter_design as fd
from matplotlib import pyplot as plt

# Specification for our filter
Wp = 0.412   # Cutoff frequency 
Ws = 0.270   # Stop frequency 
Rp = 0.1     # passband maximum loss (gpass)
As = 60      # stoppand min attenuation (gstop)

Filters = {'moving' : (), 'butter' : ()}

# butter and chebyshev 1 filter design
Filters['butter'] = fd.iirdesign(Ws, Wp, Rp, As, ftype='butter')
Filters['chebyshev'] = fd.iirdesign(Ws, Wp, Rp, As, ftype='cheby1')

plt.plot(Filters['chebyshev'])
plt.show()
```
